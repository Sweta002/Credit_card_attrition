## Check outliers in dataset

```python
def outliers_iqr(ys):
    quartile_1, quartile_3 = np.percentile(ys, [25, 75])
    iqr = quartile_3 - quartile_1
    lower_bound = quartile_1 - (iqr * 1.5)
    upper_bound = quartile_3 + (iqr * 1.5)
    sns.boxplot(x = ys.index, data=ys)
```

```python
outliers_iqr(dataset['Total_TransCount'])
```

```python
plt.figure(figsize=(15,10))
plt.subplot(5,2,1)
outliers_iqr(dataset['Customer_Age'])
plt.subplot(5,2,2)
outliers_iqr(dataset['Dependent_count'])
plt.subplot(5,2,3)
outliers_iqr(dataset['Months_on_book'])
plt.subplot(5,2,4)
outliers_iqr(dataset['Total_Relationship_Count'])
plt.subplot(5,2,5)
outliers_iqr(dataset['Months_Inactive_12_mon'])
plt.subplot(5,2,6)
outliers_iqr(dataset['Contacts_Count_12_mon'])
plt.subplot(5,2,7)
outliers_iqr(dataset['Credit_Limit'])
plt.subplot(5,2,8)
outliers_iqr(dataset['Total_Revolving_Bal'])
plt.subplot(5,2,9)
outliers_iqr(dataset['Total_TransAmount'])
plt.subplot(5,2,10)
outliers_iqr(dataset['Total_TransCount'])
plt.savefig("Box")
```

## Treating outliers

```python
def fill_outliers(a):
    floor = a.quantile(0.05)
    cap = a.quantile(0.99)
    a.fillna(cap)
    a.fillna(floor)
```

```python
fill_outliers(dataset.Months_on_book)
fill_outliers(dataset.Months_Inactive_12_mon)
fill_outliers(dataset.Contacts_Count_12_mon)
fill_outliers(dataset.Credit_Limit)
fill_outliers(dataset.Total_TransAmount)
fill_outliers(dataset.Total_TransCount)
```

## Checking Multi-collinearity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def Calc_vif (data):
    vif = pd.DataFrame()
    vif['features'] = data.columns
    vif['VIF_value'] = [variance_inflation_factor(data.values, i) for i in range(data.shape[1])]
    return (vif)
```
